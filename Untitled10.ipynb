```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
# Анализ данных

# A. Загрузка и первичный обзор
file_path = 'google_books_dataset.csv'
df = pd.read_csv(file_path)

print('Вывод первых пяти строк для проверки')
print(df.head())


print('Вывод последних пяти строк для проверки')
print(df.tail())

print('Структура данных')
print(df.info())
print(df.dtypes)


# B. Обработка данных


print(df.isnull())
print('Кол-во пропусков')
print(df.isnull().sum())

df['authors']= df['authors'].fillna('Unknown Author')
df['publisher'] =df['publisher'].fillna('Unknown Publisher')
df['description']=df['description'].fillna('No Description Available')
df['categories']=df['categories'].fillna('Unknown Category')
df['title']=df['title'].fillna('Unknown Title')
df['published_date']=df['published_date'].fillna('Unknown Data')
df['thumbnail']=df['thumbnail'].fillna('No thumbnail')
df['isbn_13']=df['isbn_13'].fillna(False)
df['isbn_10']=df['isbn_10'].fillna(False)

columns_to_drop_na = ['subtitle', 'average_rating', 'list_price', 'currency']
df.drop(columns=columns_to_drop_na, inplace = True)

median_page_count = df['page_count'].median()
df['page_count']=df['page_count'].fillna(median_page_count)

print('Кол-во пропусков после заполнения')
print(df.isnull().sum())

print('Дубликаты')
d = df.duplicated()
print(d)
print('Кол-во дублирующихся строк')
duplicate_rows = df.duplicated().sum()

print(duplicate_rows)
print('Дубликатов нет и удалять не нужно ничего')



print("Метод z-score")
mean_val1 = df['page_count'].mean()
std_val1 = df['page_count'].std()
z_score_threshold = 3

df['z_score_' + 'page_count'] = np.abs((df['page_count'] - mean_val1) / std_val1)
outliers_zscore = df[df['z_score_' + 'page_count'] > z_score_threshold]

print(f"Среднее '{'page_count'}': {mean_val1:.2f}")
print(f"Стандартное отклонение '{'page_count'}': {std_val1:.2f}")
print(f"Количество выбросов (Z-score > {z_score_threshold}): {len(outliers_zscore)}")


print("Примеры выбросов по Z-score:")
print(outliers_zscore[['page_count', 'z_score_' + 'page_count']].head())


# Пример обработки: Замена выбросов по Z-score на граничные значения
upper_limit_zscore = mean_val1 + z_score_threshold * std_val1
lower_limit_zscore = mean_val1 - z_score_threshold * std_val1

# Применяем замену только если есть выбросы
df['page_count'] = np.where(df['page_count'] > upper_limit_zscore, upper_limit_zscore, df['page_count'])
df['page_count'] = np.where(df['page_count'] < lower_limit_zscore, lower_limit_zscore, df['page_count'])
print(f"Выбросы по Z-score заменены на граничные значения [{lower_limit_zscore:.2f}, {upper_limit_zscore:.2f}].")


# Удаляем временный столбец Z-score
df = df.drop(columns=['z_score_' + 'page_count'])
print("Boxplot (визуальный и IQR):")

plt.figure(figsize=(5, 4))
sns.boxplot(y=df['page_count'])
plt.title(f'Boxplot для столбца "{'page_count'}"')
plt.ylabel('page_count')
plt.show()



print("Метод z-score")
mean_val2 = df['ratings_count'].mean()
std_val2 = df['ratings_count'].std()
z_score_threshold = 3

df['z_score_' + 'ratings_count'] = np.abs((df['ratings_count'] - mean_val2) / std_val2)
outliers_zscore = df[df['z_score_' + 'ratings_count'] > z_score_threshold]

print(f"Среднее '{'ratings_count'}': {mean_val2:.2f}")
print(f"Стандартное отклонение '{'ratings_count'}': {std_val2:.2f}")
print(f"Количество выбросов (Z-score > {z_score_threshold}): {len(outliers_zscore)}")


print("Примеры выбросов по Z-score:")
print(outliers_zscore[['ratings_count', 'z_score_' + 'ratings_count']].head())


# Пример обработки: Замена выбросов по Z-score на граничные значения
upper_limit_zscore = mean_val2 + z_score_threshold * std_val2
lower_limit_zscore = mean_val2 - z_score_threshold * std_val2

# Применяем замену только если есть выбросы
df['ratings_count'] = np.where(df['ratings_count'] > upper_limit_zscore, upper_limit_zscore, df['ratings_count'])
df['ratings_count'] = np.where(df['ratings_count'] < lower_limit_zscore, lower_limit_zscore, df['ratings_count'])
print(f"Выбросы по Z-score заменены на граничные значения [{lower_limit_zscore:.2f}, {upper_limit_zscore:.2f}].")


# Удаляем временный столбец Z-score
df = df.drop(columns=['z_score_' + 'ratings_count'])
print("Boxplot (визуальный и IQR):")

plt.figure(figsize=(5, 4))
sns.boxplot(y=df['ratings_count'])
plt.title(f'Boxplot для столбца "{'ratings_count'}"')
plt.ylabel('ratings_count')
plt.show()

# Статистика данных:
print('Описательная статистика')
print(df.describe())


print('Подсчет уникальных значений')
print(df.nunique())

print('Подсчет частоты authors:')
print(df['authors'].value_counts())
print('Подсчет частоты publisher:')
print(df['publisher'].value_counts())


print('Корреляционный анализ')
ct = df.select_dtypes(include=[np.number]).corr()
print(ct)



# Визуализация данных:
sns.set_style('whitegrid')
sns.set_context('talk')

df['published_date'] = pd.to_datetime(df['published_date'], errors='coerce')
df['published_year'] = df['published_date'].dt.year
books_per_year = df['published_year'].value_counts().sort_index()





plt.figure(figsize=(12, 6))
sns.histplot(df['page_count'], bins=50, kde=True, color='forestgreen', edgecolor='black')
plt.title('Распределение количества страниц', fontsize=18, fontweight='bold')
plt.xlabel('Количество страниц', fontsize=14)
plt.ylabel('Частота', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()


plt.figure(figsize=(14, 7))
sns.lineplot(x=books_per_year.index, y=books_per_year.values, color='forestgreen', linewidth=2.5)
plt.title('Динамика количества опубликованных книг по годам', fontsize=18, fontweight='bold')
plt.xlabel('Год публикации', fontsize=14)
plt.ylabel('Количество опубликованных книг', fontsize=14)
plt.yticks(fontsize=12)
plt.grid(True, linestyle=':', alpha=0.6)
plt.tight_layout()
plt.show()


plt.figure(figsize=(14, 8))
top_categories = df['categories'].value_counts().head(10)
sns.barplot(x=top_categories.values, y=top_categories.index, hue=top_categories.index,
            palette='viridis', dodge=False, legend=False, edgecolor='black', linewidth=0.8)
plt.title('Топ 10 самых популярных категорий книг', fontsize=18, fontweight='bold')
plt.xlabel('Количество книг', fontsize=14)
plt.ylabel('Категория', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()




plt.figure(figsize=(16, 7))

plt.subplot(1, 2, 1)
sns.boxplot(y=df['page_count'], color='lightcoral')
plt.title('Boxplot количества страниц', fontsize=16, fontweight='bold')
plt.ylabel('Количество страниц', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.subplot(1, 2, 2)
sns.boxplot(y=df['ratings_count'], color='lightseagreen')
plt.title('Boxplot количества рейтингов', fontsize=16, fontweight='bold')
plt.ylabel('Количество рейтингов', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.tight_layout()
plt.show()


plt.figure(figsize=(8, 6))
sns.heatmap(df[['page_count', 'ratings_count']].corr(), annot=True, cmap='viridis', fmt=".2f", cbar_kws={'label': 'Коэффициент корреляции'})
plt.title('Heatmap корреляции количества страниц и количества рейтингов', fontweight='bold')
plt.show()

# Группировки и агрегации
print('Среднее количество страниц и рейтингов по категориям')
mean_by_category = df.groupby('categories')[['page_count', 'ratings_count']].mean().sort_values(by='page_count', ascending=False)
print(mean_by_category)

print('Сумма количества страниц по годам публикации')
sum_by_year = df.groupby('published_year')['page_count'].sum().sort_index()
print(sum_by_year)

print('Сводная таблица: Среднее количество страниц по категории и году публикации')
pivot_table_ = pd.pivot_table(df, values='page_count', index='categories', columns='published_year', aggfunc='mean')
print(pivot_table_)
```
